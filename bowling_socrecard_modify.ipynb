```python
import pandas as pd
```

```python
df = pd.read_csv("bowling_scorecard_data.csv")
```

```python
df.head()
```

```python
df=df.drop(["odi_year","Match_id"],axis=1)
```

```python
print(df.dtypes)
```

```python
df['4s'] = pd.to_numeric(df['4s'], errors='coerce')
df['4s'] = df['4s'].fillna(0).astype('int64')
```

```python
df['6s'] = pd.to_numeric(df['6s'], errors='coerce')
df['6s'] = df['6s'].fillna(0).astype('int64')
```

```python
print(df.dtypes)
```

```python
df["balls_balled"] = (df["O"]//1)*6 + round(df["O"]%1,1)*10
```

```python
df["extras"] = df["WD"]+df["NB"]
```

```python
df["boundaries"] = df["4s"]+df["6s"]
```

```python
df.head()
```

```python
df = df.groupby(["BOWLING","team"]).agg({'W': 'sum', 'ECON': 'mean','R':'sum',"balls_balled":"sum","extras":"sum","boundaries":"sum","M":'sum'}).reset_index()
```

```python
print(df.dtypes)
```

```python
df
```

```python
df["normalized_W"] = ((df["W"]-df["W"].min())
                      /(df["W"].max()-df["W"].min()))*10
```

```python
df["normalized_ECON"] = (1-
                         ((df["ECON"]-df["ECON"].min())
                            /(df["ECON"].max()-df["ECON"].min()))
                         )*10
```

```python
df["normalized_extras"] = (1-
                           ((df["extras"]-df["extras"].min())/
                              (df["extras"].max()-df["extras"].min()))
                             )*10
```

```python
df["boundary_percentage"] = df["boundaries"]/df["balls_balled"]
```

```python
df["normalized_boundary_perc"] = (1-((df["boundary_percentage"]-df["boundary_percentage"].min())
                                     /(df["boundary_percentage"].max()-df["boundary_percentage"].min()))
                                    )*10
```

```python
df
```

```python
df.to_csv("Normalized_bowling_scorecard.csv")
```
